```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

group = {} #建立一個空字典

#抓取在part_list_2屬性內的h3資料
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text #時間
    title = d.find_all('a')[-1].text.replace('\u3000', '')  #標題
    tag = d.find('em').text #類別，find為收尋HTML內的文件
    #tag = d.find_all('em')[0].text
    #tag = d.find_all('em')[-1].text
#   tag = d.select('.tag')[0].text #類別，select為收尋CSS內的文素
    group.setdefault(tag, [])
    group[tag].append({
        'title': title,
        'date': date
    })
    
print(group)
```

```python
data = []
for i in group:
    print(i, len(group[i]))
    d = {}
    d['tag'] = i
    d['count'] = len(group[i])
    data.append(d)

for d in sorted(data, key=lambda d: d['count'])[::-1]:
    print(d)
```
