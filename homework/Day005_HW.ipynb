```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests,json
url = " https://www.dcard.tw/_api/forums/pet/posts?popular=true"
fl = requests.get(url)
file = json.loads(fl.text)
print(len(file))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for ls in file:
    
    title =ls["title"]
    time = ls["createdAt"]
    comment = ls["commentCount"]
    like = ls["likeCount"]

    print("標題:",title,"\n","貼文時間",time,"\n","留言人數",comment,"\n","按讚人數",like)
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
#熱門
com_count = 0
like_count = 0
for i in file:
    com_count += i["commentCount"]
    like_count += i["likeCount"]

avgcomment = com_count/len(file)
avglike = like_count/len(file)
print("熱門文章")
print("平均留言人數 :%.2f"%(avgcomment))
print("平均按讚人數 :%.2f"%(avglike))
```
