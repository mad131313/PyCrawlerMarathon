```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests,json
import pandas as pd
url = " https://www.dcard.tw/_api/forums/pet/posts?popular=true"
fl = requests.get(url)
file = json.loads(fl.text)
print(len(file))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for ls in file:
    
    title =ls["title"]
    time = ls["createdAt"]
    comment = ls["commentCount"]
    like = ls["likeCount"]

    print("æ¨™é¡Œ:",title,"\n","è²¼æ–‡æ™‚é–“",time,"\n","ç•™è¨€äººæ•¸",comment,"\n","æŒ‰è®šäººæ•¸",like)
```

```python
#ä½¿ç”¨Pandaså–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
#ä½¿ç”¨Pandasè¨ˆç®—ç†±é–€çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
title,time,comment,like = [], [], [], []
for ls in file:
    title.append(ls["title"])
    time.append(ls["createdAt"])
    comment.append(ls["commentCount"])
    like.append(ls["likeCount"])
data = {"title":title,"time":time,"comment":comment,"like":like}
aa = pd.DataFrame(data)

print("è¨ˆç®—ç†±é–€çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€")
print("å¹³å‡ç•™è¨€äººæ•¸",aa["comment"].mean())
print("å¹³å‡æŒ‰è®šäººæ•¸",aa["like"].mean())
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
#ç†±é–€
for i in file:
    com_count += i["commentCount"]
    like_count += i["likeCount"]

avgcomment = com_count/len(file)
avglike = like_count/len(file)

print("å¹³å‡ç•™è¨€äººæ•¸: %.2f"%(avgcomment))
print("å¹³å‡æŒ‰è®šäººæ•¸ :%.2f"%(avglike))
```
